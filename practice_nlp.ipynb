```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet

def get_synonyms_antonyms(word):
    synonyms = set()
    antonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            # Add synonyms
            synonyms.add(lemma.name())
            # Check and add antonyms
            if lemma.antonyms():
                for ant in lemma.antonyms():
                    antonyms.add(ant.name())

    return synonyms, antonyms

# Word to analyze
word = "active"

synonyms, antonyms = get_synonyms_antonyms(word)

print(f"Synonyms for '{word}': {synonyms}")
print(f"Antonyms for '{word}': {antonyms}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,Input,Embedding,LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
src=['hello','how are you','good morning']
tgt=['hola','como estas','buenos dias']
tgt_in=['<start>'+t for t in tgt]
tgt_out=[t+'<end'for t in tgt]
def tok(x):
  t=Tokenizer()
  t.fit_on_texts(x)
  return pad_sequences(t.texts_to_sequences(x)),t
x,s_tok=tok(src)
y_in,t_tok=tok(tgt_in)
y_out,_=tok(tgt_out)
ei=Input(shape=(None,))
_,h,c=LSTM(256,return_state=True)(Embedding(len(t_tok.word_index)+1,128)(ei))
di=Input(shape=(None,))
do = LSTM(256, return_sequences=True)(
        Embedding(len(t_tok.word_index)+1, 128)(di),
        initial_state=[h, c]
     )
out=Dense(len(t_tok.word_index)+1,'softmax')(do)
m=Model([ei,di],out)
m.compile('adam','sparse_categorical_crossentropy')
m.summary()
m.fit([x,y_in],np.expand_dims(y_out,-1),epochs=100,verbose=0)
```

```python
import nltk
from nltk.corpus import wordnet
nltk.download('wordnet')
def saw(w):
  s=set()
  a=set()
  for syn in wordnet.synsets(w):
    for l in syn.lemmas():
      s.add(l.name())
      if l.antonyms():
        for an in l.antonyms():
          a.add(an.name)
  return a,s
word='activate'
a,s=saw(word)
print(f'{s}\n{a}')
```
