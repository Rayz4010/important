```python
# @title Default title text
import nltk
from nltk import word_tokenize, FreqDist
from nltk.util import ngrams
from collections import defaultdict
import math
nltk.download('punkt')
def compute_ngram_probabilities(text, n):
  """ Compute n-gram probabilities for a given text."""
  tokens = word_tokenize(text.lower()) # Tokenize and convert to lowercase
  n_grams = list(ngrams(tokens, n))
  if not n_grams:
    return {}
  total_ngrams = len(n_grams)
  ngram_freq = FreqDist(n_grams)
  # Probability calculation
  ngram_probabilities = {}
  for ngram, freq in ngram_freq.items():
    ngram_probabilities[ngram] = freq / total_ngrams

  return ngram_probabilities

def display_ngram_probabilities(text, max_n=3):
  """ Compute and display unigram, bigram, and trigram probabilities."""
  for n in range(1, max_n + 1):
    print(f"\n{n}-Gram Probabilities:")
    ngram_probs = compute_ngram_probabilities(text, n)
    for ngram, prob in ngram_probs.items():
      print(f"{ngram}: {prob:.4f}")

# Example sentences
text = "The quick brown fox jumps over the lazy dog. The fox is quick and clever."
display_ngram_probabilities(text)
```

```python
import nltk
from collections import defaultdict
from nltk.probability import FreqDist

# Training data (Japanese words)
train_data = [
("楽しい, カップル, 愛, 愛", "コメディ"),
("速い, 怒り, 撃つ", "アクション"),
("カップル, 飛ぶ, 速い, 楽しい, 楽しい", "コメディ"),
("怒り, 撃つ, 撃つ, 楽しい", "アクション"),
("飛ぶ, 速い, 撃つ, 愛", "アクション")
]

# Tokenize and prepare vocabulary
nltk.download('punkt')

category_words = defaultdict(list)
all_words = set()
category_counts = defaultdict(int)

for text, category in train_data:
    words = text.split(', ')
    category_words[category].extend(words)
    all_words.update(words)
    category_counts[category] += 1

total_docs = len(train_data)
total_vocab = len(all_words)

# Compute prior probabilities
priors = {category: category_counts[category] / total_docs for category in category_counts}

# Compute likelihoods with add-1 smoothing
likelihoods = {}

for category in category_words:
    freq_dist = FreqDist(category_words[category])
    total_category_words = len(category_words[category])

    likelihoods[category] = {
        word: (freq_dist[word] + 1) / (total_category_words + total_vocab)
        for word in all_words
    }

# Classify new document D
D = "速い, カップル, 撃つ, 飛ぶ"
D_words = D.split(', ')

posteriors = {}

for category in category_words:
    posterior = priors[category]

    for word in D_words:
        posterior *= likelihoods[category].get(
            word, 1 / (len(category_words[category]) + total_vocab)
        )

    posteriors[category] = posterior

# Determine the most likely category
predicted_category = max(posteriors, key=posteriors.get)

print("Predicted category for D:", predicted_category)

print("Posterior probabilities:")
for category, prob in posteriors.items():
    print(f"{category}: {prob}")
```

```python
from collections import defaultdict

data = [
    ("fun couple love love", "comedy"),
    ("fast furious shoot", "action"),
    ("couple fly fast fun fun", "comedy"),
    ("furious shoot shoot fun", "action"),
    ("fly fast shoot love", "action")
]

cw, cc, vocab = defaultdict(list), defaultdict(int), set()

for text, c in data:
    w = text.split()
    cw[c] += w
    vocab |= set(w)
    cc[c] += 1

V, N = len(vocab), len(data)
priors = {c: cc[c]/N for c in cc}

# likelihoods with smoothing
lik = {
    c: {w: (cw[c].count(w)+1)/(len(cw[c])+V) for w in vocab}
    for c in cw
}

D = "fast couple shoot fly".split()
post = {
    c: priors[c] *
       __import__('functools').reduce(
           lambda a, w: a * lik[c].get(w, 1/(len(cw[c])+V)), D, 1)
    for c in cw
}

print("Prediction:", max(post, key=post.get))
print("Posteriors:", post)
```

```python
import nltk
from nltk.corpus import brown, inaugural, reuters, udhr
from nltk.probability import ConditionalFreqDist
from nltk.tag import UnigramTagger
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
from collections import Counter
import string

# Download necessary datasets
nltk.download("brown")
nltk.download("inaugural")
nltk.download("reuters")
nltk.download("udhr")
nltk.download("averaged_perceptron_tagger")

nltk.download("punkt")
nltk.download("universal_tagset")
nltk.download("wordnet")

# Exploring Various Corpora
def explore_corpus():
    print("Brown Corpus Categories:", brown.categories())
    print("First 100 words from Brown Corpus:", brown.words()[:100])
    print("First 100 words from Inaugural Corpus:", inaugural.words()[:100])
    print("First 100 words from Reuters Corpus:", reuters.words()[:100])
    print("UDHR Corpus Samples:", udhr.words('English-Latin1')[:100])

# Creating Custom Corpus
def create_custom_corpus():
    corpus_text = """Natural Language Processing is amazing. It is a subset of Artificial
Intelligence."""
    with open("custom_corpus.txt", "w") as f:
        f.write(corpus_text)
    with open("custom_corpus.txt", "r") as f:
        text = f.read()
    print("Custom Corpus Text:", text)

# Conditional Frequency Distribution
def conditional_frequency():
    cfd = ConditionalFreqDist((genre, word) for genre in brown.categories() for word in
                              brown.words(categories=genre))
    print("Most common words in news category:", cfd["news"].most_common(10))

# Studying Tagged Corpora
def tagged_corpora():
    print("Tagged words from Brown Corpus:", brown.tagged_words(tagset='universal')[:20])

# Finding Most Frequent Noun Tags
def most_frequent_noun_tags():
    tagged_words = brown.tagged_words(tagset='universal')
    noun_tags = [word for word, tag in tagged_words if tag == 'NOUN']
    noun_counts = Counter(noun_tags)
    print("Most frequent nouns:", noun_counts.most_common(10))

# Mapping Words to Properties Using Python Dictionaries
def map_words_to_properties():
    words = ["car", "apple", "computer", "python"]
    word_properties = {word: wn.synsets(word) for word in words}
    for word, properties in word_properties.items():
        print(f"{word}: {[synset.definition() for synset in properties][:2]}")

# Rule-based and Unigram Tagger
def rule_based_tagger():
    tagged_sents = brown.tagged_sents(categories='news')
    unigram_tagger = UnigramTagger(tagged_sents[:5000])
    test_sent = word_tokenize("The quick brown fox jumps over the lazy dog")
    print("Tagged sentence:", unigram_tagger.tag(test_sent))

# Finding different words from plain text without spaces
def split_text_without_spaces(text, corpus_words):
    found_words = []
    start = 0
    while start < len(text):
        for end in range(len(text), start, -1):
            if text[start:end] in corpus_words:
                found_words.append(text[start:end])
                start = end # Move start to the end of the found word to avoid overlapping and improve efficiency
                break
        else: # If no word was found, advance by one character to avoid infinite loop
            start += 1
    return found_words

if __name__ == "__main__":
    explore_corpus()
    create_custom_corpus()
    conditional_frequency()
    tagged_corpora()
    most_frequent_noun_tags()
    map_words_to_properties()
    rule_based_tagger()

    # Example text without spaces
    corpus_words = set(brown.words())
    text = "thequickbrownfoxjumpsoverthelazydog"
    found_words = split_text_without_spaces(text, corpus_words)
    print("Segmented Words:", found_words)
```

```python
import nltk
nltk.download('wordnet')
#nltk.download('omw-1.4')
from nltk.corpus import wordnet

def get_synonyms_antonyms(word):
    synonyms = set()
    antonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            # Add synonyms
            synonyms.add(lemma.name())
            # Check and add antonyms
            if lemma.antonyms():
                for ant in lemma.antonyms():
                    antonyms.add(ant.name())

    return synonyms, antonyms

# Word to analyze
word = "active"

synonyms, antonyms = get_synonyms_antonyms(word)

print(f"Synonyms for '{word}': {synonyms}")
print(f"Antonyms for '{word}': {antonyms}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample data - usually you'd load real parallel sentences
source_sentences = ['hello', 'how are you', 'good morning']
target_sentences = ['hola', 'cómo estás', 'buenos días'] # Spanish example

# Add start and end tokens to target
target_sentences_input = ['<start> ' + sentence for sentence in target_sentences]
target_sentences_output = [sentence + ' <end>' for sentence in target_sentences]

# Tokenize
def tokenize(sentences):
  tokenizer = Tokenizer(filters='')
  tokenizer.fit_on_texts(sentences)
  tensor = tokenizer.texts_to_sequences(sentences)
  tensor = pad_sequences(tensor, padding='post')
  return tensor, tokenizer

source_tensor, source_tokenizer = tokenize(source_sentences)
target_input_tensor, target_tokenizer = tokenize(target_sentences_input)

target_output_tensor, _ = tokenize(target_sentences_output)

source_vocab_size = len(source_tokenizer.word_index) + 1
target_vocab_size = len(target_tokenizer.word_index) + 1

embedding_dim = 256
units = 512

# Encoder
encoder_inputs = Input(shape=(None,))
enc_emb = Embedding(source_vocab_size, embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(units, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(None,))
dec_emb_layer = Embedding(target_vocab_size, embedding_dim)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = LSTM(units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = Dense(target_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
model.summary()

# Train

model.fit([source_tensor, target_input_tensor],np.expand_dims(target_output_tensor,-1),batch_size=2,epochs=100)
```

```python
import numpy as np
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Data
src = ['hello', 'how are you', 'good morning']
tgt = ['hola', 'cómo estás', 'buenos días']

# Add tokens
tgt_in  = ['<start> ' + t for t in tgt]
tgt_out = [t + ' <end>' for t in tgt]

# Tokenize
def tok(x):
    t = Tokenizer()
    t.fit_on_texts(x)
    return pad_sequences(t.texts_to_sequences(x)), t

X, s_tok = tok(src)
Y_in, t_tok = tok(tgt_in)
Y_out, _ = tok(tgt_out)

# Model
ei = Input(shape=(None,))
_, h, c = LSTM(256, return_state=True)(Embedding(len(s_tok.word_index)+1, 128)(ei))

di = Input(shape=(None,))
do = LSTM(256, return_sequences=True)(
        Embedding(len(t_tok.word_index)+1, 128)(di),
        initial_state=[h, c]
     )
out = Dense(len(t_tok.word_index)+1,'softmax')(do)

model = Model([ei, di], out)
model.compile('adam', 'sparse_categorical_crossentropy')
model.summary()
# Train
model.fit([X, Y_in], np.expand_dims(Y_out, -1), epochs=100, verbose=0)
```
